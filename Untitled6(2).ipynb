```python
# @title Task 1:Creating the Dataset
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import torch
from PIL import Image


```

```python
# @title 1.1)Load Image
image_path = '/content/sample_data/rainbow1-1.jpg'
image = Image.open(image_path).convert('RGB')  # Convert to RGB to simplify data
```

```python
# @title 1.2)Extract Coordinates and RGB Values
# Convert to NumPy array
image_array = np.array(image)

# Get dimensions
height, width, _ = image_array.shape

# Create x and y coordinate arrays
x_coords = np.tile(np.arange(width), height)
y_coords = np.repeat(np.arange(height), width)

# Reshape the image array to extract RGB values
rgb_values = image_array.reshape(-1, 3)
```

```python
# @title 1.3)Create a Pandas DataFrame:
# Combine into a Pandas DataFrame
data = pd.DataFrame({
    'x': x_coords,
    'y': y_coords,
    'R': rgb_values[:, 0],
    'G': rgb_values[:, 1],
    'B': rgb_values[:, 2]
})

data1=data.copy()
```

```python
# @title 1.4)Inspect the DataFrame
# Display the first 10 rows of the DataFrame
print(data.head(10))
```

**Challanges**

**Solution**
I used NumPy's reshape and tile/repeat functions to flatten the RGB data and generate corresponding x and y coordinates systematically.

**Challange**


*   For high-resolution images, the amount of data (width × height × 5 columns) can become very large, potentially leading to memory issues.

**Solution**



*   I used efficient memory management of NumPy and Pandas for handling large datasets.

**Challange**
* The code needs to adapt to images of different dimensions or formats without breaking.

**Solution**
* The use of .shape to dynamically extract the height and width ensures that the code works for images of any resolution.

```python

# @title 2.2) Analyze noise in the dataset
dark_noise_threshold = 30
bright_noise_threshold = 230

# Identify dark and bright noise
dark_noise = data[(data['R'] < dark_noise_threshold) & (data['G'] < dark_noise_threshold) & (data['B'] < dark_noise_threshold)]
bright_noise = data[(data['R'] > bright_noise_threshold) & (data['G'] > bright_noise_threshold) & (data['B'] > bright_noise_threshold)]

print(f"Number of dark noise pixels: {len(dark_noise)}")
print(f"Number of bright noise pixels: {len(bright_noise)}")

```

```python
# @title 2.3) Filter the dataset to remove noise
filtered_data = data[
    ~((data['R'] < dark_noise_threshold) & (data['G'] < dark_noise_threshold) & (data['B'] < dark_noise_threshold)) &
    ~((data['R'] > bright_noise_threshold) & (data['G'] > bright_noise_threshold) & (data['B'] > bright_noise_threshold))
]


```

```python
# @title 2.3) Reconstruct and display the cleaned image
cleaned_array = np.zeros_like(image_array)
for index, row in filtered_data.iterrows():
    cleaned_array[row['y'], row['x']] = [row['R'], row['G'], row['B']]

plt.figure(figsize=(8, 6))
plt.imshow(cleaned_array)
plt.title("Cleaned Image")
plt.axis('off')
plt.show()
```

```python
#@title Task 3:KMeans Clustering with scikit-learn
```

```python
#@title 3.1)Standardize the Data:
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['x', 'y', 'R', 'G', 'B']])
```

```python
#@title 3.2) Perform KMeans Clustering
n_clusters = 6  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
data['cluster'] = kmeans.fit_predict(scaled_data)
```

```python
#@title 3.3) Add Cluster Labels to the DataFrame:
distinct_colors = list(mcolors.TABLEAU_COLORS.values())[:n_clusters]
cluster_color_map = {i: tuple(int(255 * c) for c in mcolors.to_rgb(distinct_colors[i])) for i in range(n_clusters)}
```

```python
#@title 3.4) Visualize the Clustered Data:

clustered_image = np.zeros((height, width, 3), dtype=np.uint8)

for index, row in data.iterrows():
    clustered_image[int(row['y']), int(row['x'])] = cluster_color_map[row['cluster']]

# Display the clustered image
plt.figure(figsize=(10, 8))
plt.imshow(clustered_image)

plt.axis('off')
plt.show()
```

```python
# @title 4.1)Prepare the Data:
import torch # Import the torch module

# Convert to PyTorch tensor
data_tensor = torch.tensor(data1[['x', 'y', 'R', 'G', 'B']].values, dtype=torch.float32)
data_tensor = (data_tensor - data_tensor.mean(dim=0)) / data_tensor.std(dim=0)  # Normalize

```

```python
# @title 4.2)Initialize Centroids:
# Initialize centroids
num_clusters = 8
initial_indices = torch.randperm(data_tensor.size(0))[:num_clusters]
centroids = data_tensor[initial_indices]
```

```python
# @title 4.3)Implement the Clustering Algorithm:
# Implement clustering algorithm
num_iterations = 20
cluster_labels = torch.zeros(data_tensor.size(0), dtype=torch.long)

for iteration in range(num_iterations):
    distances = torch.cdist(data_tensor, centroids)
    cluster_labels = torch.argmin(distances, dim=1)

    new_centroids = torch.zeros_like(centroids)
    for k in range(num_clusters):
        cluster_points = data_tensor[cluster_labels == k]
        if len(cluster_points) > 0:
            new_centroids[k] = cluster_points.mean(dim=0)
        else:
            new_centroids[k] = data_tensor[torch.randint(0, data_tensor.size(0), (1,))]

    if torch.allclose(centroids, new_centroids, atol=1e-4):
        break

    centroids = new_centroids


```

```python
# @title 4.4)Add Cluster Labels to the DataFrame:
# Add cluster labels to the DataFrame
data1['cluster'] = cluster_labels.numpy()

# Map clusters to colors for visualization
distinct_colors = list(mcolors.TABLEAU_COLORS.values())[:num_clusters]
cluster_color_map = {i: tuple(int(255 * c) for c in mcolors.to_rgb(distinct_colors[i])) for i in range(num_clusters)}

# Get original image dimensions
height, width, _ = image_array.shape

# Calculate downscaled dimensions
new_height = int(height * 0.5)
new_width = int(width * 0.5)


clustered_downscaled_image = np.zeros((new_height, new_width, 3), dtype=np.uint8)
for index, row in data1.iterrows():
    # Adjust coordinates for downscaling
    downscaled_y = int(row['y'] * 0.5)
    downscaled_x = int(row['x'] * 0.5)

    # Check if downscaled coordinates are within bounds
    if 0 <= downscaled_y < new_height and 0 <= downscaled_x < new_width:
        clustered_downscaled_image[downscaled_y, downscaled_x] = cluster_color_map[row['cluster']]
```

```python
# @title 4.5)Visualize the Clusters:


# Display the clustered image
plt.figure(figsize=(10, 8))
plt.imshow(clustered_downscaled_image)

plt.axis('off')
plt.show()
```
